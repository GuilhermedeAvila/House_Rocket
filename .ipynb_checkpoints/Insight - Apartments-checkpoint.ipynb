## 0.1 Helper Fuctions


```python
def jupyter_settings():
    %matplotlib inline
    %pylab inline
    warnings.filterwarnings('ignore')
    plt.style.use( 'bmh' )
    plt.rcParams['figure.figsize'] = [20, 10]
    plt.rcParams['font.size'] = 18
    display( HTML( '<style>.container { width:100% !important; }</style>') )
    pd.options.display.max_columns = None
    pd.options.display.max_rows = None
    pd.set_option( 'display.expand_frame_repr', False )
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    sns.set()
    
    
def desc_statistics(numerical_attributes):
    ct1 = pd.DataFrame( numerical_attributes.apply( np.mean ) ).T
    ct2 = pd.DataFrame( numerical_attributes.apply( np.median) ).T

    #Dispersion - std, min, max, ramge, skew, kurtosis

    d1 = pd.DataFrame( numerical_attributes.apply( np.std ) ).T
    d2 = pd.DataFrame( numerical_attributes.apply( min ) ).T
    d3 = pd.DataFrame( numerical_attributes.apply( max ) ).T
    d4 = pd.DataFrame( numerical_attributes.apply( lambda x: x.max() - x.min() ) ).T
    d5 = pd.DataFrame( numerical_attributes.apply( lambda x: x.skew() ) ).T
    d6 = pd.DataFrame( numerical_attributes.apply( lambda x: x.kurtosis() ) ).T

    # concatenar

    stats = pd.concat([ d2, d3, d4, ct1, ct2, d1, d5, d6] ).T.reset_index()
    stats.columns = ['attributes', 'min', 'max', 'range', 'mean', 'median', 'std','skew', 'kurtosis']
    
    return stats
    
def get_data( path ):
    data = pd.read_csv( path )
    #Suprimir notificacao cientifica
    np.set_printoptions(suppress=True)
    pd.set_option('display.float_format','{:.2f}'.format)
    pd.set_option('display.float_format', lambda x: '%.2f' % x)
    data['date']= pd.to_datetime(data['date'], format = '%Y-%m-%d')
    data['query'] = data[['lat', 'long']].apply( lambda x: str( x['lat'] ) + ',' + str( x['long'] ), axis=1)
    return data

def media_lineplot(dataset, col1, col2, x, y, xlabel, ylabel, title):
    aux1 = dataset[[col1, col2]].groupby(col1).mean().reset_index();
    ax = sns.lineplot( x=x, y=y, data=aux1);
    ax.set(xlabel=xlabel, ylabel=ylabel);
    ax.set(title=title);
       
    return ax;

def mediana_barplot(dataset, col1, col2, x, y, xlabel, ylabel, title):
    aux1 = dataset[[col1, col2]].groupby(col1).median().reset_index();
    ax = sns.barplot( x=x, y=y, data=aux1);
    ax.set(xlabel=xlabel, ylabel=ylabel);
    ax.set(title=title);
    
    return ax;

def media_barplot(dataset, col1, col2, x, y, xlabel, ylabel, title):
    aux1 = dataset[[col1, col2]].groupby(col1).median().reset_index();
    ax = sns.barplot( x=x, y=y, data=aux1);
    ax.set(xlabel=xlabel, ylabel=ylabel);
    ax.set(title=title);
    
    return ax;

def count_barplot(dataset, col1, col2, x, y, xlabel, ylabel, title):
    aux1 = dataset[[col1, col2]].groupby(col1).count().reset_index();
    ax = sns.barplot( x=x, y=y, data=aux1);
    ax.set(xlabel=xlabel, ylabel=ylabel);
    ax.set(title=title);
    
    return ax;
```

```python
df1 = df.copy()
```

### 2.1.1 Hipótese 1
Imóveis que possuem vista para água, são 30% mais caros, na média.


#### Falsa! 

```python
preco_media_waterfront_1 = np.round( df1.loc[df1['waterfront'] == 1, 'price'].mean(), 2)
print('Preço médio de imóveis que possuem vista para a água: ${}'.format(preco_media_waterfront_1))
preco_media_waterfront_0 = np.round( df1.loc[df1['waterfront'] < 1, 'price'].mean(), 2)
print('Preço médio de imóveis que não possuem vista para a água: ${}'.format(preco_media_waterfront_0))
preco_media_imoveis = np.round( df1['price'].mean(), 2 )
print('Preço médio dos imóveis: ${}'.format(preco_media_imoveis) )

#validando a hipótese
hipotese1 = ((preco_media_waterfront_1 - preco_media_imoveis ) / preco_media_imoveis) * 100
print('\nImóveis que possuem vista para a água, são na verdade {:.2f}% mais caros, na média!'.format(hipotese1))
```

### 2.1.2 Hipótese 2
Imóveis com data de construção menor que 1955, são 50% mais baratos, na média.

#### Falsa!

```python
#Verificando a hipótese
h2_ = preco_media_1955.loc[0, 'preco_medio']
h2_1 = preco_media_1955.loc[1, 'preco_medio']
h2_f = np.round(((h2_ - h2_1) / h2_) * 100, 2)
print(f'Imóveis com ano de construção menor que 1955 são na verdade {h2_f}% mais baratos, na média.')

```

### 2.1.3 Hipótese 3
Imóveis sem porão possuem em média o sqrt_lot 50% maior do que com porão. (Desconsiderando o sqft_lot15)

#### Falsa!

```python
sem_porao = df1.loc[df1['sqft_basement'] == 0, 'sqft_lot'].mean()
com_porao = df1.loc[df1['sqft_basement'] > 0, 'sqft_lot'].mean()

#validando a hipótese
hipotese3 = ((sem_porao - com_porao) / com_porao ) * 100

print('Imóveis sem porão possuem em média o sqrt_lot(terreno) {:.2f}%. maior do que com porão.'.format(hipotese3))
```

# 4.0 Recomendações de compra e venda

## 4.1 Quais são os imóveis que a House Rocket deveria comprar?
- Agrupar os zipcodes
- Calcular a mediana dos preços/m2,
- Os imóveis com condição igual ou acima de 3
- Imóveis que forem abaixo do preço da mediana correspondente ao seu zipcode serão recomendados para a compra.

```python
# carregando os dados
path = 'kc_house_data.csv'
df = get_data( path )
# criando uma coluna com o valor dos pes quadrados de cada imovel
df['price/sqft'] = df['price'] / df['sqft_lot']

# agrupando por zipcode e tirando a mediana
zip_code = df[['zipcode', 'price/sqft']].groupby('zipcode').median().reset_index()
zip_code.columns = ['zipcode', 'price_zipcode']
# concatenando por pela mediana do preco por zipcode
df91 = pd.merge(df, zip_code, on='zipcode', how='inner')

# criando coluna de recomendação e definindo se compra ou não o imóvel
df91['recommendation'] = df91[['price/sqft', 'price_zipcode', 'condition']].apply(
                                                                                 lambda x: 'buy' if ( x['price/sqft'] <= x['price_zipcode']) &
                                                                                 ( x['condition'] >= 3 ) else 'not buy', axis=1
                                                                                 )

df92 = df91.loc[df91['recommendation'] == 'buy', :].copy()
print(f'A quantidade de casas recomendadas a compra: {len(df92)}\n\n')
df92.head()

```

## 4.2 Uma vez a casa comprada, qual o melhor momento para vendê-las e por qual preço?
- Agrupar por zipcodes e estações do ano, tirar a mediana deste grupo.
- Condições de venda: Se o preço da compra for maior que a mediana da região + sazonalidade: O preço da venda será igual ao preço da compra + 10%. Se o preço da compra for menor que a mediana da região + sazonalidade. O preço da venda será igual ao preço da compra + 30%.

Para entender a relação das estações do ano com o preço de venda do imóvel, vamos considerar a data do registro do imóvel no conjunto de dados.

Ao agrupar os imóveis por região e estação do ano em que foi anunciada, será possível comparar os preços medianos da região e por sazonalidade.

Por se tratar de imóveis a venda nos EUA, as estações do ano acontecem nas seguintes datas:


- Inverno: 21 de dezembro a 20 de março;
- Primavera: 21 de março a 20 de junho;
- Verão: 21 de junho a 22 de setembro;
- Outono: 23 de setembro a 20 de dezembro.

Para efeito de análise vou considerar as estações separadas por meses, logo:

- Inverno: jan a mar;
- Primavera: abr a jun;
- Verão: jul a set;
- Outono: out a dez.

```python
# Criando as estacoes
df92.insert(2, 'month', df92['date'].dt.month)


df92['season'] = df92['month'].apply(lambda x: 
                                   'winter' if 1 <= x <= 3 else 
                                   'spring' if 4 <= x <= 6 else 
                                   'summer' if 7 <= x <= 9 else 
                                   'fall')

df_zip_season = df92[['price/sqft', 'zipcode', 'season']].groupby(['zipcode', 'season']).median().reset_index()
df_zip_season.rename({'price/sqft': 'price/sqft_season'}, axis=1, inplace=True)

df93 = pd.merge(df92, df_zip_season, on=['zipcode', 'season'])

# Nessa parte sera calculado os valores de venda.
df93['sell_price'] = df93[['price/sqft', 'price/sqft_season', 'sqft_lot']].apply( lambda x: ( x['price/sqft'] * x['sqft_lot'] * 1.3 ) 
                                                                        if x['price/sqft'] <= x['price/sqft_season'] else 
                                                                        ( x['price/sqft'] * x['sqft_lot'] * 1.1 ), axis=1 )
df93.head()
```

```python
# Salvando csv com recomendações de compra
df93.to_csv('casas_para_compra.csv')
```

# 5.0 - Conclusão

```python
# Resultado Financeiro
conc = df93.copy()
conc['gain'] = conc[['price', 'sell_price']].apply( lambda x: x['sell_price'] - x['price'], axis = 1)
soma_conc = np.round(conc['gain'].sum(),2)
soma_precos = conc['price'].sum()
percentagem_conc = np.round((soma_conc/soma_precos) * 100, 2)

print('O potencial total do lucro obtido com as vendas recomendadas: ${:.2f} , resultando em {}%. de lucro'.format(soma_conc, percentagem_conc))
```
